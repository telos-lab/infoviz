## Example Data: Gaussian Mixture

```python
rng = np.random.default_rng(seed=42)
df = pd.DataFrame(np.vstack([rng.normal(size=(10000,2), loc=-1, scale=1),
                           rng.normal(size=(10000,2), loc=1.5, scale=0.75),
                             rng.normal(size=(1000,2), loc=0, scale=1.5)
                          ]), columns=["x","y"])
```

### Original scatter plot

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.scatterplot(df, x="x", y="y", color=cc.glasbey_hv[1])
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_original.pdf", transparent=True, bbox_inches="tight")
```

### Remove marker fill

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.scatterplot(df, x="x", y="y", color=cc.glasbey_hv[1], marker=r"$\circ$", ec=cc.glasbey_hv[1])
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_nofill.pdf", transparent=True, bbox_inches="tight")
```

### Change marker symbol

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.scatterplot(df, x="x", y="y", color=cc.glasbey_hv[1], marker=r"+", lw=1.5)
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_marker-symbol.pdf", transparent=True, bbox_inches="tight")
```

### Change marker size

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.scatterplot(df, x="x", y="y", color=cc.glasbey_hv[1], marker="o", s=5, lw=0)
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_marker-size.pdf", transparent=True, bbox_inches="tight")
```

### Increase marker transparency

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.scatterplot(df, x="x", y="y", color=cc.glasbey_hv[1], marker="o", s=10, lw=0, alpha=0.2)
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_marker-alpha.pdf", transparent=True, bbox_inches="tight")
```

### Add KDE lines

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.scatterplot(df, x="x", y="y", color=cc.glasbey_hv[1], marker="o", s=10, lw=0, alpha=0.2)
sns.kdeplot(df, x="x", y="y", color="black", fill=False, alpha=0.5)
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_kde-lines.pdf", transparent=True, bbox_inches="tight")
```

### Replace scatter by KDE

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.kdeplot(df, x="x", y="y", color=cc.glasbey_hv[1], fill=True)
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_kde-filled.pdf", transparent=True, bbox_inches="tight")
```

### Replace scatter by 2-dimensional histogram

```python
fig, ax = plt.subplots(1,1, figsize=(6,6))
sns.histplot(df, x="x", y="y", color=cc.glasbey_hv[1], fill=True)
ax.set_xticks(np.arange(-4,5,4))
ax.set_yticks(np.arange(-4,5,4))
ax.set_xlabel("")
ax.set_ylabel("")
sns.despine(fig, trim=True)
plt.savefig("../figures/overplotting_gaussian-scatter_hist2d.pdf", transparent=True, bbox_inches="tight")
```
