```python
from infoviz.utils import load_df
```

```python
df = load_df("iris")
```

```python
target_names = list(df.drop_duplicates(subset="y_name")[["y","y_name"]].sort_values("y").y_name)
```

```python
fig, ax = plt.subplots(1,1,figsize=(9,6))
pd.plotting.parallel_coordinates(
    df[df.columns[:-1]], "y", color=cc.glasbey_hv[:3], ax=ax
) 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, target_names)
ax.set_ylim(0,8)
plt.tight_layout()
plt.savefig("../figures/parallel_coordinates-iris.pdf", bbox_inches="tight", transparent=True)
```

```python
alt_cols = list(reversed(['sepal length (cm)',   'petal length (cm)','sepal width (cm)','petal width (cm)','y']))
fig, ax = plt.subplots(1,1,figsize=(9,6))
pd.plotting.parallel_coordinates(
    df[alt_cols], "y", color=cc.glasbey_hv[:3], ax=ax
) 
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, target_names)
ax.set_ylim(0,8)
plt.tight_layout()
plt.savefig("../figures/parallel_coordinates-iris-alt.pdf", bbox_inches="tight", transparent=True)
```
